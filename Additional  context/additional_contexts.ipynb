

# ADDING ADDITIONAL QUESTIONS




```python
import json
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define input and output file paths
input_file = '/content/drive/MyDrive/drivingcontexts_questions.json'  # Change if in a subfolder
output_file = '/content/drive/MyDrive/updated_drivingcontexts_questions.json'

# Define new questions
additional_questions = [
    "Is the traffic light green?",
    "Is the road blocked due to an accident?",
    "Is there a sharp turn ahead?",
    "Are there animals on the road?",
    "Is there a speed bump ahead?",
    "Is the vehicle approaching a railroad crossing?",
    "Are there traffic cones or barricades?",
    "Is the driver approaching a toll booth?"
]

def process_large_json(input_file, output_file):
    try:
        with open(input_file, 'r') as file:
            data = json.load(file)  # Load JSON data

        # Ensure "questions" key exists
        if "questions" not in data:
            raise KeyError("The key 'questions' is missing in the JSON file.")

        updated_questions = []

        # Iterate over existing questions and add additional questions after every 24 questions
        for i, question in enumerate(data["questions"]):
            updated_questions.append(question)
            if (i + 1) % 24 == 0:  # After every 24 questions
                new_entries = [
                    {"image_id": question["image_id"], "question": q, "question_id": f"{question['image_id']}_{24+j}"}
                    for j, q in enumerate(additional_questions)
                ]
                updated_questions.extend(new_entries)

        # Write updated data back to JSON
        with open(output_file, 'w') as outfile:
            json.dump({"questions": updated_questions}, outfile, indent=4)

        print("Updated JSON file saved successfully.")
    except Exception as e:
        print(f"Error processing file: {e}")

# Call the function
process_large_json(input_file, output_file)
```
